```python
size = 1000
yy = np.random.rand(size)
xx = np.arange(size)
np.sort(yy) 
count = 0
check = 0
while(count<size):
    yy[count] += count
    hold = (random.uniform(-2.5,3.5)) 
    yy[count] = hold
    check +=1
    count +=6

randSpot = int(size-size/10)
op = size / 25
while(randSpot < (size-size/10)+op):
    yy[randSpot] += 18
    randSpot+=1
    
origGraph = [1] * size
x = 0
while (x < size) :
    origGraph[x] = yy[x]
    x += 1
```

```python
fig = go.Figure(data=go.Scatter(x=xx, y=yy, mode='markers'))

fig.show()
```

```python
from scipy.optimize import curve_fit

timeStart = time.time()

def func(x, a, b):
    return a * x + b

p, pcov = curve_fit(func, xx, yy)
y_exp_outliers = func(xx, *p)


fig = go.Figure(data = go.Scatter(x = xx, y=yy,
    name='Original Plot', mode='markers'))
fig.add_trace(go.Scatter(
    x=xx,
    y=y_exp_outliers,
    mode='markers',
    marker=dict(
        color='black',
        size=5
    ),
    name='Linear'
))
fig.update_layout(
      xaxis=dict(
        title='Time (s)'
    ),
    yaxis=dict(
        title='Position in X'
    ))
fig.show()
```

```python
def func(x, a, b):
    return a * x + b

count = 0
while(count<size):
    diff = y_exp_outliers[count] - yy[count]
    if abs(diff) > 1.5:
        yy[count] = y_exp_outliers[count]
    count +=1
    
count = 0    
while(count<50):   
    p, pcov = curve_fit(func, xx, yy)
    y_exp_holder = func(xx, *p)
    insideCount = 0
    while(insideCount<size):
        diff = y_exp_holder[count] - yy[count]
        if abs(diff) > 0.75:
            yy[count] = y_exp_holder[count]         
        insideCount += 1    
        
    count +=1
```

```python
from math import exp, expm1
import math
iterations = 0
bestFit = []
bestLine = p
sampleValue = int(size / 5)
sampleSize = sampleValue
bestOutliers = 100
bestA = 0
bestB = 0

timeStart = time.time()

x_temp = np.arange(sampleSize)

def func(x, a, b):
    return a * x + b

i = 0
N = sampleValue # amount of samples taken

while(i < N):
    outliers = 0
    samples = 0
    set = [1] * sampleSize
    
    while(samples < sampleSize): # picks random points for micro graph
        samp = int((random.uniform(0,size))) 
        set[samples] = origGraph[samp]
        samples += 1
    
    p, pcov = curve_fit(func, x_temp, set) # fits a y = mx+b line to the selected points
    mini_y_exp = func(x_temp, *p)
    
    count = 0
    while(count<sampleSize):
        diff = mini_y_exp[count] - set[count]
        if abs(diff) > 0.3:
            outliers += 1
        count +=1
    if outliers < bestOutliers:
        bestOutliers = outliers
        bestA = p[0]
        bestB = p[1]
    
    i += 1
    
timeEnd = time.time()
timeElapsedRansac = timeEnd - timeStart
```

```python
print(p) # a * x + b

fig = go.Figure()
fig.add_trace(go.Scatter(x = xx, y=origGraph,
    name='Original Plot', mode='markers'))
fig.add_trace(go.Scatter(
    x=xx,
    y = (bestA * xx + bestB),
    mode='markers',
    marker=dict(
        color='black',
        size=5
    ),
    name='RANSAC Algorithm'
))
fig.add_trace(go.Scatter(
    x=xx,
    y=y_exp,
    mode='markers',
    marker=dict(
        color='red',
        size=5
    ),
    name='Trivial Solution'
))
fig.add_trace(go.Scatter(
    x=xx,
    y=y_exp_outliers,
    mode='markers',
    marker=dict(
        color='cyan',
        size=2
    ),
    name='No change linear regression'
))

fig.show()

print("Time elapsed in seconds for RANSAC =", timeElapsedRansac, "\n")	
print("Time elapsed in seconds for Trivial Solution=", timeElapsed, "\n")	
```
